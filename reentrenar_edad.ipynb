# Reentrenamiento del modelo de edad

Este notebook usa el flujo limpio del repositorio para reentrenar un único modelo de regresión de edad.

Solo necesitas cambiar `MAX_IMAGES` y volver a ejecutar la celda de entrenamiento.

```python
from pathlib import Path
import sys

REPO_ROOT = Path("lab02_profesor/Lab02-ML-2026-01-main").resolve()
if str(REPO_ROOT) not in sys.path:
    sys.path.insert(0, str(REPO_ROOT))

from src.config import LabConfig
from src.training import run_age_training_workflow
```

```python
# Cambia solo este valor para volver a entrenar con más o menos fotos.
DATASET_DIR = Path("UTKFace")
OUTPUT_DIR = REPO_ROOT / "artifacts_notebook"
MAX_IMAGES = 2500
IMAGE_SIZE = (25, 25)
TEST_SIZE = 0.20
RANDOM_STATE = 42
PCA_COMPONENTS = (30, 50, 80, 100, 150, 200)

config = LabConfig(
    dataset_dir=DATASET_DIR,
    output_dir=OUTPUT_DIR,
    image_size=IMAGE_SIZE,
    test_size=TEST_SIZE,
    random_state=RANDOM_STATE,
    pca_components=PCA_COMPONENTS,
    max_images=MAX_IMAGES,
)
config
```

```python
summary = run_age_training_workflow(config)
summary
```

```python
print("Modelo:", summary["model_path"])
print("Metricas:", summary["metrics_path"])
print("Figura:", summary["figure_path"])
print("Mejor configuracion:", summary["best_params"])
print("Metricas:", summary["metrics"])
```
